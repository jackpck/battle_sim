```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for i in [4,5,6,7,8]:
    df = pd.read_csv('./data/health1_%.1f_reserve_fraction_%.3f.csv'%(i,0), index_col=0)
    plt.plot(df.Nteam1, df.favor, label=i)
plt.plot(df.Nteam1,[0]*len(df.Nteam1),'k--')
plt.legend(loc='best')
plt.xlabel('team 1 size')
plt.ylabel('Favor larger team   |   Favor smaller team')
plt.title('winning trend with different starting health')
plt.show()
```
